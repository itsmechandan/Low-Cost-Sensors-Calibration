```python
import pandas as pd
df = pd.read_excel('/content/4SENSORS(NaCl).xlsx')
```

```python
df
```

```python
x = df.iloc[:,[1] + list(range(3, 7))]

y = df.iloc[:,list(range(1,3))]

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)

from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=100, random_state=42)

rf.fit(x_train, y_train)
```

```python
y_pred = rf.predict(x_test)
y_pred_df = pd.DataFrame(y_pred,columns=y_test.columns)
y_pred_df['Time'] =df['Time']
y_pred_df
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2_ = r2_score(y_test, y_pred)

print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("R-squared:", r2_)
```

```python
y_pred_full = rf.predict(x)
y_pred_df_full = pd.DataFrame(y_pred_full,columns=y_test.columns)
y_pred_df_full['Time'] =df['Time']
y_pred_df_full
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


mae = mean_absolute_error(y, y_pred_full)
mse = mean_squared_error(y, y_pred_full)
r2_ = r2_score(y, y_pred_full)

print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("R-squared:", r2_)
```

```python
y_pred_df_full.to_excel("4SENSORSCOMBINED_PRED(Nacl).xlsx", index=False)
```
